```python
# Import the pandas library for data manipulation
import pandas as pd
```

```python
# Load the main dataset from an Excel file
df = pd.read_excel('datasets/premiums.xlsx')
# Display the first 5 rows to verify it loaded correctly
df.head()
```

```python
df.shape
```

```python
# Filter the dataset to include only records where Age is greater than 25
df_rest = df[df['Age'] > 25]
# Check the shape (number of rows and columns) of the filtered dataset
df_rest.shape
```

```python
df_rest.head()
```

```python
# Filter the dataset to include only records where Age is less than or equals to 25
df_young = df[(df['Age'] < 25) | (df['Age'] == 25)]
# Check the shape of this young demographic dataset
df_young.shape
```

```python
df_young.head()
```

```python
# Save the dataset of people older than 25 to a new Excel file
df_rest.to_excel('datasets/premiums_rest.xlsx', index=False)
# Save the dataset of people younger than 25 to a new Excel file
df_young.to_excel('datasets/premiums_young.xlsx', index=False)
```
